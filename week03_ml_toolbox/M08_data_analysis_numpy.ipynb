```python
arr = np.random.rand(2,3)
print(arr)
print(arr.ndim)
print(arr.dtype)
print(arr.shape)
print(arr.size)
```

```python
arr = np.random.randint(1, 10, (2,3))
print(arr)
print(arr.ndim)
print(arr.dtype)
print(arr.shape)
print(arr.size)
```

```python
arr = np.random.uniform(50, 100, (3,4))
print(arr)
print(arr.ndim)
print(arr.dtype)
print(arr.shape)
print(arr.size)
```

```python
arr = np.arange(1, 10, 1)
arr = arr.reshape(3,3)
print(arr)
print(arr.ndim)
print(arr.dtype)
print(arr.shape)
print(arr.size)
```

```python

arr = np.linspace(1, 10, 10)

print(arr)
print(arr.ndim)
print(arr.dtype)
print(arr.shape)
print(arr.size)
```

```python

arr = np.logspace(0, 4, 5, base=2)

print(arr)
```
