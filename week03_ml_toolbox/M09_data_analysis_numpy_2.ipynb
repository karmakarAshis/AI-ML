```python
# inplace sorting

x = np.array([10, 8, 16, 100])
z = x.copy()

print(z)

z.sort()
print(z)

```

```python
# copy sorting

print(x)
sort_arr = np.sort(x)
print(sort_arr)

print(x)
```

```python
# 2d array
mat = np.array([[10, 4, 7], [6, 3, 9]])

# horizon, axix=1
hor_sort = np.sort(mat, axis=1)
print(hor_sort)
```

```python
# vertical, axix=0
ver_sort = np.sort(mat, axis=0)
print(ver_sort)
```

```python
# np.where (condition, x, y) if true x, otherwiae replace with y
x = np.array([10, 8, 16, 100])
print(x)


index = np.where(x==16)
print(index)
```

```python
index = np.where(x>10)
print(index)
```

```python
x = np.array([10, 100, 8, 16, 100])

maximum_value_index = np.argmax(x)
minimum_value_index = np.argmin(x)

print(maximum_value_index)
print(minimum_value_index)

```

```python
# counting in array
a = np.random.randint(1, 100, size=(100,))
print(a)
```

```python
value_greater_than_60 = np.count_nonzero(a>70)

print(value_greater_than_60)
```

```python
unique_value, count = np.unique(a, return_counts=True)
print(unique_value, count)
```
